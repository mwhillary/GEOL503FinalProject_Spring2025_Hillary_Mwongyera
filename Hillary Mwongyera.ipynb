## Making plots using matplotlib.pyplot

_1. Import the matplotlib.pyplot module and then make a plot of the function $y = e^x$ (aka the exponential function) for $x$ values from -5 to 5.  Plot the function as a red line._

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating an array of x values from -5 to 5
x_values = np.linspace(-5, 5)

# Calculating the corresponding y values for y = e^x
y_values = np.exp(x_values)

# Plotting the function as a red line
plt.plot(x_values, y_values, 'r-')

# Adding labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = e^x')

# Display the plot
plt.show()
```

_2. Make a new figure and set of axes, and use them to plot the functions $y = \sqrt{x}$ and $y = x + 3$ on the same axes. Plot the functions in two different colors over the range $x = [0, 5]$._

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating x values from 0 to 5
x_values = np.linspace(0, 5, 100)

# Defining the functions
y_sqrt_x = np.sqrt(x_values)
y_x_plus_3 = x_values + 3

# Creating a new figure and axes
fig, ax = plt.subplots()

# Plotting the functions in different colors
ax.plot(x_values, y_sqrt_x, 'b-', label=r'$y = \sqrt{x}$')  # Blue line for y = √x
ax.plot(x_values, y_x_plus_3, 'g-', label=r'$y = x + 3$')    # Green line for y = x + 3

# Adding labels, title, and legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Plot of y = √x and y = x + 3')
ax.legend()

# Display the plot
plt.show()
```
